# Import Libraries

```python
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
import numpy as np
from tensorflow.keras.optimizers import Adam
```

# Load Model

```python
# Load the model without the optimizer
model = load_model('model.h5')
```

# Load Data & Predict

```python
# Load the image
img_path = "/Users/ciputrawangsa/Downloads/test_inf.jpg"
img = image.load_img(img_path, target_size=(224, 224))  # Resize to match the input shape
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0  # Normalize if required

# Make prediction
prediction = model.predict(img_array)

# Convert integer label to class name
class_labels = ['Heart', 'Oblong', 'Oval', 'Round', 'Square']  # Update with actual class names
predicted_class = np.argmax(prediction, axis=1)
predicted_class_name = class_labels[predicted_class[0]]

# Plot the image with predicted class
plt.imshow(img)
plt.title(f"Predicted: {predicted_class_name}")
plt.axis('off')  # Turn off axis
plt.show()

print("Predicted class name:", predicted_class_name)
```
